## Data Cleaning

Filling NaN values with their column means and round the means in order to keep the structure same.
Additionally, dropping rows which has at least one NaN value after filling.

```
playStore_df = playStore_df.fillna(round(playStore_df.mean(),1))
playStore_df = playStore_df.dropna(how = "any")
playStore_df.isnull().sum()
```

Duplicate rows have been eliminated from the data with "drop_duplicates" function.

```
playStore_df.drop_duplicates(inplace = True)
```

Unneccessery columns containing non interpretable data, have been eliminated from the data with drop function.

```
playStore_df.columns
```

```
playStore_df = playStore_df.drop(columns = ["Last Updated", "Current Ver", "Android Ver"])
```

```
playStore_df.columns
```

The column which contains the information of the size for each app has been converted to float type, by getting rid of letters abd multiplying the integers with the value of corresponding size unit.

```
def SizeTypeCast(df):
  size = str(df["Size"])
  
  if "M" in size:
    size = size.replace("M", "")
    size = float(size)
    size *= 1024
  elif "k" in size:
    size = size.replace("k", "")
    size = float(size)
  else:
    size = -1.0

  return size

copied_df = playStore_df.copy()
copied_df["Size"] = playStore_df.apply(SizeTypeCast, axis = 1)
copied_df
```

Deleting the "+" signs in "Installs" column, converting values to integer type.

```
def InstallsToInt(df):
  installs  = str(df["Installs"])
  installs = installs.replace("+", "")
  installs = installs.replace(",", "")
  installs = int(installs)
  return installs

copied_df["Installs"] = copied_df.apply(InstallsToInt, axis = 1)
copied_df
```

Removing "$" signs in "Price" column, converting price values into float type.

```
def FixPrice(df):
  price  = str(df["Price"])
  price = price.replace("$", "")
  price = float(price)
  return price

copied_df["Price"] = copied_df.apply(FixPrice, axis = 1)
copied_df.dtypes
```

Converting the type of "Reviews" column to into from string.

```
copied_df["Reviews"] = copied_df["Reviews"].astype(int)
copied_df.dtypes
```

## Machine Learning

```
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score, roc_curve
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing
```

```
le = preprocessing.LabelEncoder() 
df_y = copied_df[["Installs"]]
df_x = copied_df.drop(["Installs", "App"], axis = 1)

for i in ["Category", "Type", "Content Rating", "Genres"]: 
  le.fit(copied_df[i]) 
  df_x[i] = list(le.transform(copied_df[i])) 

X_train, X_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.25, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.25, random_state=42)

print(X_train.shape, X_val.shape, X_test.shape, y_train.shape, y_val.shape, y_test.shape)
```

### Decision Tree

The desicion tree method has been used and the depth of the three has been modified in order to reach the optimal accuracy rate

```
max_depth_list = list(range(1,15))
tree_acc_list = []
for depth in max_depth_list:
  model_dt = tree.DecisionTreeClassifier(random_state=42, max_depth = depth)
  model_dt.fit(X_train, y_train)

  dt_predictions = model_dt.predict(X_val)
  dt_acc = accuracy_score(y_val, dt_predictions)
  tree_acc_list.append(dt_acc)

plt.figure(figsize=(12, 6))
plt.plot(max_depth_list, tree_acc_list, color='black', linestyle='dashed', marker='o',markerfacecolor='brown', markersize=10)
plt.plot([6], [0.542], color = "purple", marker = "o", markersize = 10)
plt.annotate("Best Accuracy on Validation Data", xy=(6, 0.532), xytext=(7, 0.46), arrowprops=dict(facecolor='black', shrink=0.1))
plt.annotate("Best Accuracy on Test Data with max_depth 6", xy=(6, 0.542), xytext=(8, 0.535), arrowprops=dict(facecolor='black', shrink=0.1))
plt.xlabel("Tree Depth")
plt.ylabel("Accuracy scores")
plt.title("Accuracy Scores of Decision Trees")
plt.grid()
plt.show()

max_score = max(tree_acc_list)
max_index = tree_acc_list.index(max_score)
ideal_depth = max_depth_list[max_index]

print("With validation data")
print("Maximum accuracy score has occurred with depth ", ideal_depth, ", and it is ", format(max_score, ".3f"), ".", sep = "")

model_dt_6 = tree.DecisionTreeClassifier(random_state=42, max_depth = 6)
model_dt_6.fit(X_train, y_train)
dt_predictions = model_dt_6.predict(X_test)
dt_acc_test = accuracy_score(y_test, dt_predictions)

print("\n\nWith test data")
print("Accuracy score with depth ", ideal_depth, " is ", format(dt_acc_test, ".3f"), ".", sep = "")
```

### Random Forest

Random forest method has been used, the tree number and maximum depth of trees has been changed according to reach the optimal accuracy score.

```
estimator_list = list(range(64, 129, 2))
accuracy_list_maxDepthGiven = []
accuracy_list_noMaxDepth = []
for estimator in estimator_list:
  model_rf = RandomForestClassifier(n_estimators= estimator, random_state=42, max_depth= 6)
  model_rf.fit(X_train, y_train.values.ravel())

  rf_predictions = model_rf.predict(X_val)
  rf_acc = accuracy_score(y_val, rf_predictions)
  accuracy_list_maxDepthGiven.append(rf_acc)

  model2_rf = RandomForestClassifier(n_estimators= estimator, random_state=42)
  model2_rf.fit(X_train, y_train.values.ravel())

  rf_predictions = model2_rf.predict(X_val)
  rf_acc2 = accuracy_score(y_val, rf_predictions)
  accuracy_list_noMaxDepth.append(rf_acc2)

plt.figure(figsize = (16,8))
plt.plot(estimator_list, accuracy_list_maxDepthGiven, color='green', linestyle='-', marker='o', markerfacecolor='black', markersize=8, label = "max_depth given as 6")
plt.plot(estimator_list, accuracy_list_noMaxDepth, color= "red", linestyle= "-", marker= "o", markerfacecolor='black', markersize=8, label = "no max_depth parameter given")
plt.plot([114],[0.55], color = "blue", markersize= 11, marker = "o")
plt.annotate("Best Accuracy\non Validation Data\nwith given depth 6", xy=(128, 0.506), xytext=(120, 0.515), arrowprops=dict(facecolor='green', shrink=0.1))
plt.annotate("Best Accuracy\non Validation Data\nwith no given depth", xy=(114, 0.543), xytext=(114, 0.530), arrowprops=dict(facecolor='green', shrink=0.1))
plt.annotate("Best Accuracy\non Test Data", xy=(114, 0.55), xytext=(100, 0.55), arrowprops=dict(facecolor='green', shrink=0.1))
plt.xlabel("Number of estimators")
plt.ylabel("Accuracy scores")
plt.title("Accuracy Scores of Random Forests")
plt.legend()
plt.grid()
plt.show()

max_score = max(accuracy_list_maxDepthGiven)
max_index = accuracy_list_maxDepthGiven.index(max_score)
ideal_estimator_number = estimator_list[max_index]

max_score_noDepth = max(accuracy_list_noMaxDepth)
max_index_noDepth = accuracy_list_noMaxDepth.index(max_score_noDepth)
ideal_estimator_number_noDepth = estimator_list[max_index_noDepth]

print("With validation data")
print("Maximum accuracy score has occurred using ", ideal_estimator_number, " many estimators (with max_depth 6) and it is ", format(max_score, ".3f"), ".", sep = "")
print("Maximum accuracy score has occurred using ", ideal_estimator_number_noDepth, " many estimators (no specific depth given) and it is ", format(max_score_noDepth, ".3f"), ".", sep = "")
model_rf_112 = RandomForestClassifier(n_estimators= 114, random_state=42)
model_rf_112.fit(X_train, y_train.values.ravel())
rf_predictions = model_rf_112.predict(X_test)
rf_acc_test = accuracy_score(y_test, rf_predictions)

print("\n\nWith test data")
print("Accuracy score with ", ideal_estimator_number_noDepth, " many estimators (no specific depth given) is ", format(rf_acc_test, ".3f"), ".", sep = "")
```

### k-Nearest Neighbor

```
knn_accuracy_list = []
for k in range (1, 25):
    knn = KNeighborsClassifier(n_neighbors = k)
    knn.fit(X_train, y_train.values.ravel())
    y_pred = knn.predict(X_val)
    knn_accuracy_list.append(accuracy_score(y_val, y_pred))

plt.figure(figsize=(12, 6))  
plt.plot(range(1, 25), knn_accuracy_list, color='m', linestyle='dashed', marker='o', markerfacecolor='y', markersize=10, label = "Accuracy scores wrt 'k'")
plt.annotate("Best\nAccuracy\non Validation\nData", xy=(8, 0.425), xytext=(8, 0.40), arrowprops=dict(facecolor='m', shrink=0.1))
plt.title('Accuracy of the Validation w/ k-Value')
plt.xlabel('k-Value')
plt.ylabel('Accuracy')
plt.legend()
plt.grid()
plt.show()

print("With validation data")
print("Maximum accuracy score has occurred with k = ", knn_accuracy_list.index(max(knn_accuracy_list))+1, " it is ", format(max(knn_accuracy_list), ".3f"), ".", sep = "")

knn_9 = KNeighborsClassifier(n_neighbors=4)
knn_9.fit(X_train, y_train.values.ravel())
y_pred = knn_9.predict(X_test)

print("\n\nWith test data")
print("Accuracy of the test with k = 8 is ", format(accuracy_score(y_test, y_pred),".3f"), ".", sep = "")
```
